```python
import pandas as pd  # To read data
# To perform computate eg. exponential, log, as well as to generate ransom noise
import numpy as np
# to perform non-linear curve fitting (i.e to compute an optimized coefficients)
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt  # To displayd ata for visualization
from sklearn.metrics import r2_score
```

```python
# Create the data set
# Når x var 5.1 ble y 8.5....osv
x = np.array([5.1, 10.2, 170.3, 340.6, 510.9, 1021.8])
y = np.array([8.5, 9.6, 19.2, 25.6, 32.1, 45.9])

# Nå må vi visualisere for å se hvilken type funksjon vi har med å gjøre
# I hans ord:
# To decide the possible mapping function, we need to display

plt.scatter(x, y)
# Logarithmic passer (fordi vi ikke har 0.0)
# Power law 
```

```python
# Define the non-linear function

def Power_Law(x, a, b): 
    return a*x**b

# Let us find the optimized parameters that gives best fit model
# popt = optimized properties

popt,_ = curve_fit(Power_Law, x, y)

# Let us print the optimized a and b
a, b =  popt
print(popt[0], popt[1])
print(f'a={a}, b={b}') 

```

```python
# Predict

y_model = Power_Law(x, a, b)

# Plot

plt.scatter(x,y)

plt.plot(x, y_model, '--', color='r')
```

```python
# Nå har vi en model
# Nå skal vi utføre Model Performance analysis

print('R2:', r2_score(y, y_model))
```

```python
# Nå skal vi prøve en annen modell (Hershy berkley noe og Bingham plastic noe)
# Define the non-linear function

def Hershy_Berkley(x, a, b, c):
    return a*x**b + c



# Let us find the optimized parameters that gives best fit model
# popt = optimized properties
popt, _ = curve_fit(Hershy_Berkley, x, y)

# Let us print the optimized a and b and c
a, b, c = popt
print(popt[0], popt[1], popt[2])
print(f'a={a}, b={b}, c={c}')

# Predict

y_model = Hershy_Berkley(x, a, b, c)

# Plot

plt.scatter(x, y)

plt.plot(x, y_model, '--', color='r')

# Nå har vi en model
# Nå skal vi utføre Model Performance analysis

print('R2:', r2_score(y, y_model))
```

```python
# Nå skal vi prøve en annen modell (Hershy berkley noe og Bingham plastic noe)
# Define the non-linear function

def Bingham_Plastic(x, a, b):
    return a + b*x


# Let us find the optimized parameters that gives best fit model
# popt = optimized properties
popt, _ = curve_fit(Bingham_Plastic, x, y)

# Let us print the optimized a and b and c
a, b = popt
print(popt[0], popt[1])
print(f'a={a}, b={b}')

# Predict

y_model = Bingham_Plastic(x, a, b)

# Plot

plt.scatter(x, y)

plt.plot(x, y_model, '--', color='r')

# Nå har vi en model
# Nå skal vi utføre Model Performance analysis

print('R2:', r2_score(y, y_model))
```
