```python
import numpy as np
import matplotlib.pyplot as plt

# This code is based on the simulation framework presented
# in Jan Kiusalaas book. See page 251. It can be used for single 
# and systems of ordinary differential equations with minor modifications.
# Here the Euler method is used. And the code below is Example 7.2 in the book
# where a second order differential equation is rewritten as a system 
# of two first order equations.


# If one shall solve a different problem, one typical have to adjust x,
# xStop, h,  y (initial conditions) and the F function + either remove or change the 
# exact solution. The integrate function is generic
# and do not have to be changed for a new problem. 



def main():
    x=0.0  # Start of integration
    xStop=100.0  # End of integration
    y=np.array([0.0,0.0])  #Initial conditions
    h=0.05  # Step size
    X,Y=integrate(F,x,y,xStop,h)
  #  print(Y)
    #yExact = 100.0*X-5.0*X**2+990.0*(np.exp(-0.1*X)-1.0)
    plt.plot(X,Y[:,0],'o')#yExact,'-')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.grid(True)
    plt.legend(('Numerical','Exact'))
    plt.show()
    
    
def integrate(F,x,y,xStop,h):
    
    # a general function that applies Eulers method. It finds
    # the solution from start to stop and returns the results
    # in arrays.
    
    X=[]
    Y=[]
    X.append(x)
    Y.append(y)
    
    while x<xStop:
        h=min(h,xStop-x)
        y=y+h*F(x,y)
        x=x+h
        X.append(x)
        Y.append(y)
    return np.array(X),np.array(Y)

def F(x,y):
    # User specified function that depends on the problem
    F=np.zeros(2)
    F[0]=y[1]
    F[1] = 9.81 - (0.2028/80.0)*y[1]**2
    return F
    

main()
```
